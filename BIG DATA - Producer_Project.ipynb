### Create the stream producer

```
import socket
import time
from datetime import datetime

def send():
  
  inputFile = "/FileStore/tables/Revenue_Report_IS24_FF4.csv"

  ifile  = open(inputFile, 'r') 
  i = 0
  
        
  while True:
    client_socket = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
    client_socket.setsockopt(socket.SOL_SOCKET, socket.SO_REUSEADDR, 1)
    client_socket.bind(('localhost', 9975))
    client_socket.listen(10)
    conn, addr = client_socket.accept()
    try:  
      while True:
        start = time.time()
        ifile  = open(inputFile, 'r') 
        for row in ifile.readlines() :
          print("sending: "+ row.rstrip() + "," + str(datetime.now()))
          message = row.rstrip() + ',' + str(datetime.now()) + "\n"
          message = message.encode()
          conn.send(message)
          time.sleep(5)

    except Exception as e:
        print(str(e))
        conn.close()
        client_socket.close()
        continue
    finally:
        conn.close()
        client_socket.close()
send()
```
